```python
from splinter import Browser
from bs4 import BeautifulSoup as bsoup
from urllib.request import urlopen
import requests
from splinter import Browser
import pandas as pd
import regex as re
import time
```

```python
#look at the total # of prodcuts and divide by 40 to get npages
npages = 4

pages = []

all_data = []

all_data.clear()


for i in range(1, npages):
    url = str('https://www.paragonsports.com/search?button=search&page='+str(i)+'&q=%2A&utf8=%E2%9C%93')
    #str('https://www.zillow.com/new-york-ny/'+str(i)+'_p/')
    pages.append(url)

    executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
    browser = Browser('chrome', **executable_path, headless=False)
    browser.visit(url)
    time.sleep(9)
    html = browser.html
    soup = bsoup(html, 'html.parser')
    
    data = soup.find_all('a', class_='product-summary__name-link')

    for a in data:    
        all_data.append(a)
        
    data.clear()
    browser.quit()  
 
```
